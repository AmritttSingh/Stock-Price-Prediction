```python
import pandas as pd
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC") #enables us to download price history for a single symbol 
```

```python
sp500 = sp500.history(period="max") #query all data from the very beginning when the index was created
```

```python
sp500
```

```python
sp500.index 
```

```python
sp500.plot.line(y="Close", use_index=True) 
#plot data in dataframe
#show index which is the trading days and dates - x axis
# closing price on y axis
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
#set up target - will stock go up or down

sp500["Tomorrow"] = sp500["Close"].shift(-1)
#shifting the prices to next day - tomorrow ->close
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
#predicting whether tom price > todays price
```

```python
sp500
```

```python
sp500 = sp500.loc["1950-01-01":].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
#n_estimators -> no of decision trees
#min_samples_split prevents overfitting


train = sp500.iloc[:-100] #all the rows except the last 100 rows
test = sp500.iloc[-100:] #last 100 rows
#with time series data, you cant use cross-validation, cuz u wil end up using future data to predict the past


predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds, index=test.index)
preds
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot()
#orange 0 -> predictions
#blue -> actual target
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
#one trading year -> 250 days, so we included 10 years of data here
#set->250, training model for a year
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
#percentage of days when the market actually went up
#value_counts() / no of rows 
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
#2days,1week,3months,1year,4years
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    #close price/rolling_avg
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    #no of days in the past X days when price went up
    
    new_predictors += [ratio_column,trend_column]
    
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1] #predict prob
    preds[preds>= .6] = 1 #threshold
    preds[preds < .6] = 0
    
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500,model,new_predictors)
```

```python
predictions = backtest(sp500,model,new_predictors)
```
